This project aims to **predict the probability of a road accident** based on a single image of a road scene. It analyzes the road condition, traffic congestion, and the presence of emergency vehicles such as ambulances or fire trucks.  

The project works as follows:  
1. **Takes an image of the road** as input.  
2. **Analyzes the scene** to determine the road status: normal, congested, or with an accident/emergency.  
3. **Predicts the probability of an accident** (from 0% to 100%).  
4. **Suggests appropriate action** based on the predicted probability, such as "Road is safe", "Drive carefully", or "High accident risk â€“ stop or take precautions".  

This project helps drivers and traffic authorities to **assess road risk and take preventive measures** before accidents occur.
